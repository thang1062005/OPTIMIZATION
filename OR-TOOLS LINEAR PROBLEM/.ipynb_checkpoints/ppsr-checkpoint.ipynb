```python
#PYTHON 
from ortools.linear_solver import pywraplp

def main():
    solver = pywraplp.Solver.CreateSolver('SCIP')
    
    # Input handling
    first_line = input().split()
    N = int(first_line[0])
    M = int(first_line[1])
    K = int(first_line[2])

    q = list(map(int, input().split()))
    Q = list(map(int, input().split()))
    
    d = []
    for i in range(2 * N + 2 * M + 1):
        row = list(map(int, input().split()))
        d.append(row)

    # Decision Variables
    X = {}
    for i in range(2 * N + 2 * M + 1):
        for j in range(2 * N + 2 * M + 1):
            for k in range(K):
                X[i, j, k] = solver.BoolVar(f'X[{i},{j},{k}]')

    Y = {}
    for i in range(N):
        for k in range(K):
            Y[i, k] = solver.BoolVar(f'Y[{i},{k}]')

    Z = {}
    for i in range(M):
        for k in range(K):
            Z[i, k] = solver.BoolVar(f'Z[{i},{k}]')

    T = [solver.NumVar(0, solver.infinity(), f'T[{k}]') for k in range(K)]
    D = solver.NumVar(0, solver.infinity(), 'D')

    # Objective
    solver.Minimize(D)

    # Constraint
    for k in range(K):
        solver.Add(T[k] <= D)

    for i in range(N):
        solver.Add(sum(Y[i, k] for k in range(K)) == 1)
        for k in range(K):
            solver.Add(X[i, i + N + M, k] == Y[i, k])

    for k in range(K):
        solver.Add(sum(q[i] * Z[i, k] for i in range(M)) <= Q[k])

    for i in range(M):
        solver.Add(sum(Z[i, k] for k in range(K)) == 1)

    for k in range(K):
        solver.Add(X[0, 0, k] == 1)

    for k in range(K):
        solver.Add(T[k] == sum(d[i][j] * X[i, j, k] for i in range(2 * N + 2 * M + 1) for j in range(2 * N + 2 * M + 1)))

    # Solve
    status = solver.Solve()

    # Output
    if status == pywraplp.Solver.OPTIMAL:
        print(K)
        for k in range(K):
            route = []
            current_point = 0
            visited = set([0])

            while True:
                found_next = False
                for j in range(2 * N + 2 * M + 1):
                    if solver.Value(X[current_point, j, k]) > 0.5 and j not in visited:
                        route.append(j)
                        visited.add(j)
                        current_point = j
                        found_next = True
                        break
                if not found_next:
                    break
            
            Lk = len(route) + 2
            print(Lk)
            print(f"0 {' '.join(map(str, route))} 0")
    else:
        print('No optimal solution found.')

if __name__ == "__main__":
    main()
```

```python
from ortools.linear_solver import pywraplp

def main():
    solver = pywraplp.Solver.CreateSolver('SCIP')
    
    # Input handling
    first_line = input().split()
    N = int(first_line[0])  # Number of passenger requests
    M = int(first_line[1])  # Number of parcel requests
    K = int(first_line[2])  # Number of taxis

    # Parcel request quantities
    q = list(map(int, input().split()))

    # Taxi capacities
    Q = list(map(int, input().split()))

    # Distance matrix
    d = []
    for _ in range(2 * N + 2 * M + 1):
        row = list(map(int, input().split()))
        d.append(row)

    # Decision Variables
    X = {}
    for i in range(2 * N + 2 * M + 1):
        for j in range(2 * N + 2 * M + 1):
            for k in range(K):
                X[i, j, k] = solver.BoolVar(f'X[{i},{j},{k}]')

    Y = {}
    for i in range(N):
        for k in range(K):
            Y[i, k] = solver.BoolVar(f'Y[{i},{k}]')

    Z = {}
    for i in range(M):
        for k in range(K):
            Z[i, k] = solver.BoolVar(f'Z[{i},{k}]')

    T = [solver.NumVar(0, solver.infinity(), f'T[{k}]') for k in range(K)]
    D = solver.NumVar(0, solver.infinity(), 'D')

    # Objective
    solver.Minimize(D)

    # Constraints
    for k in range(K):
        solver.Add(T[k] <= D)

    for i in range(N):
        solver.Add(sum(Y[i, k] for k in range(K)) == 1)
        for k in range(K):
            solver.Add(X[i, i + N + M, k] == Y[i, k])

    for k in range(K):
        solver.Add(sum(q[i] * Z[i, k] for i in range(M)) <= Q[k])

    for i in range(M):
        solver.Add(sum(Z[i, k] for k in range(K)) == 1)

    for k in range(K):
        solver.Add(X[0, 0, k] == 1)

    for k in range(K):
        solver.Add(T[k] == sum(d[i][j] * X[i, j, k] for i in range(2 * N + 2 * M + 1) for j in range(2 * N + 2 * M + 1)))

    # Solve
    status = solver.Solve()

    # Output
    if status == pywraplp.Solver.OPTIMAL:
        print(K)
        for k in range(K):
            route = []
            current_point = 0
            visited = set([0])

            while True:
                found_next = False
                for j in range(2 * N + 2 * M + 1):
                    if solver.Value(X[current_point, j, k]) > 0.5 and j not in visited:
                        route.append(j)
                        visited.add(j)
                        current_point = j
                        found_next = True
                        break
                if not found_next:
                    break
            
            Lk = len(route) + 2
            print(Lk)
            print(f"0 {' '.join(map(str, route))} 0")
    else:
        print('No optimal solution found.')

if __name__ == "__main__":
    main()
```
